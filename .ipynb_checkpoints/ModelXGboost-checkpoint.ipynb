## Busqueda de mejores Estimadores e HiperParametros

```python
from sklearn.metrics import accuracy_score, make_scorer
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.model_selection import train_test_split, GridSearchCV

# Definimos el modelo XGBoost para clasificación multiclase
model = xgb.XGBClassifier(objective='multi:softmax', num_class=4, random_state=42, tree_method='hist')

# Definimos un rango más acotado de estimadores
param_grid = {
    'n_estimators': [100, 200, 300, 400, 500]
}

# Configuramos una validación cruzada estratificada con menos divisiones
cv = StratifiedShuffleSplit(n_splits=3, test_size=0.2, random_state=42)

# Configuramos GridSearchCV
grid_search = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    cv=cv,
    scoring=make_scorer(accuracy_score),
    n_jobs=-1,
    verbose=2
)

# Realizamos la búsqueda
grid_search.fit(X_train, y_train)

print("Mejor número de estimadores:", grid_search.best_params_['n_estimators'])
print("Mejor puntuación:", grid_search.best_score_)

# Curva de aprendizaje (opcional, puede ser costoso computacionalmente)
results = grid_search.cv_results_
plt.figure(figsize=(10, 6))
plt.plot(param_grid['n_estimators'], results['mean_test_score'])
plt.xlabel('Número de estimadores')
plt.ylabel('Precisión media')
plt.title('Curva de aprendizaje para n_estimators')
plt.show()

# Usamos el mejor número de estimadores para ajustar otros parámetros
best_n_estimators = grid_search.best_params_['n_estimators']

# Definimos un espacio de parámetros más acotado
param_grid = {
    'max_depth': [3, 6],
    'learning_rate': [0.01, 0.1],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}

# Creamos un nuevo modelo con el mejor número de estimadores
model = xgb.XGBClassifier(objective='multi:softmax', num_class=4, n_estimators=best_n_estimators, 
                          random_state=42, tree_method='hist')

# Configuramos GridSearchCV para los otros parámetros
grid_search = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    cv=cv,
    scoring=make_scorer(accuracy_score),
    n_jobs=-1,
    verbose=2
)

# Realizamos la búsqueda
grid_search.fit(X_train, y_train)

print("Mejores parámetros:", grid_search.best_params_)
print("Mejor puntuación:", grid_search.best_score_)

# Evaluamos el mejor modelo en el conjunto de prueba
best_model = grid_search.best_estimator_
test_score = best_model.score(X_test, y_test)
print("Puntuación en el conjunto de prueba:", test_score)
```
